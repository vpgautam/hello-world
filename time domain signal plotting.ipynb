The initial downconversion of the RF frequency happens at 3.75MHz. That is, the frequency at which the signal is sampled is 3.75 MHz and the frequency of sampling is 28.8 Mega Samples per Second.
The number of data points that I have taken is 16384. And the rate of data acquision is 2.048E6.
Summarizing the above information:
Fif = 3.75 MHz.
Fs  = 28.8 MHz.
fdaq = 2.048 MHz.
Number of data points in one frame = 16384.
It does not matter at what frequency the signal is sent,a portion of the spectra (roughly around 6MHz ) is downconverted to a IF frequency of 3.75MHz.
One imporatant point to note here is that the programmes has set an offset of 1MHz while taking the samples. From the above figure, it is clear that 140 samples are required to complete one cycle.
Now, if we calculate the FFT of a multiple of 140 then the signal will be bin centered.
